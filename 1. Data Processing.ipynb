```python
import pandas as pd

# Specify the data file path
datafile_path = "2020_Data_V0.1.csv"

# Load the dataset
data = pd.read_csv(datafile_path)

# Function to check for missing and invalid data
def check_missing_and_invalid_data(df):
    invalid_data = []

    # Iterate through each column
    for column in df.columns:
        for index, value in enumerate(df[column]):
            # Check for missing data
            if pd.isnull(value):
                invalid_data.append({
                    'Row': index + 1,  # Add 1 to make row index human-readable
                    'Attribute': column,
                    'Value': 'Missing'
                })
            # Check for invalid data (example: negative numbers, empty strings)
            elif isinstance(value, str) and value.strip() == "":
                invalid_data.append({
                    'Row': index + 1,
                    'Attribute': column,
                    'Value': 'Empty String'
                })
            elif isinstance(value, (int, float)) and value < 0:
                invalid_data.append({
                    'Row': index + 1,
                    'Attribute': column,
                    'Value': value
                })

    return invalid_data

# Check the data for missing and invalid entries
invalid_entries = check_missing_and_invalid_data(data)

# Print the results
if invalid_entries:
    print("Invalid or Missing Data Found:")
    for entry in invalid_entries:
        print(f"Row {entry['Row']}, Attribute: {entry['Attribute']}, Value: {entry['Value']}")
else:
    print("No missing or invalid data found.")
```

```python
import pandas as pd

# Specify the data file path
datafile_path = "2020_Data_V0.1.csv"

# Load the dataset
data = pd.read_csv(datafile_path)

# Function to count occurrences and display results
def count_values(column_name):
    try:
        value_counts = data[column_name].value_counts()
        return pd.DataFrame({column_name: value_counts.index, 'Count': value_counts.values})
    except KeyError as e:
        print(f"Column not found: {e}")
        return None

# Count occurrences for each column
port_type_counts = count_values('Port Type')
port_number_counts = count_values('Port Number')
plug_type_counts = count_values('Plug Type')

# Display the results
if port_type_counts is not None:
    print("\nPort Type Counts:")
    print(port_type_counts)

if port_number_counts is not None:
    print("\nPort Number Counts:")
    print(port_number_counts)

if plug_type_counts is not None:
    print("\nPlug Type Counts:")
    print(plug_type_counts)
```
